## 3. Functions

- A graph of the function, $f(x) = x^2 - 4x -1$
- Determine the $x$-values where $f(x) =1$
- Determine the inverse function, $g^{-1}(x)$, of the function $y = g(x) = \frac{2x-1}{x+1}.$ Plot g(x) and $g^{-1}(x)$ on the same graph.
- Given that $\alpha(x) = 2x - 5$ and $\beta(x) = 1 - x^2$, determine the composite functions, $\alpha(\beta(x))$ and $\beta(\alpha(x))$.

```python
# plot a curve
from matplotlib import pyplot as plt
import numpy as np
def f(x):
    return x**2-4*x-1
x = np.linspace(-2, 6, 100)
plt.plot(x,f(x))
plt.xlabel("x"), plt.ylabel("f(x)")
```

```python
# determine where f(x)=1
from sympy import symbols, solve
x = symbols("x")
solve(x**2 - 4*x -2, x)
```

```python
# determine the inverse function
from sympy import symbols, solve
x, y = symbols("x y")
solve((2*x-1)/(x+1)-y, x)
```

```python
# plot g(x) and its inverse function
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 6, 50)
g = (2*x-1)/(x+1)
ginv = (x+1)/(2-x)
plt.rcParams["font.size"] = 11
plt.ylim(-10, 10)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,g, label = "g(x)")
plt.plot(x, ginv, label = "$g^{-1}(x)$")
plt.legend()
```

```python
# composition of functions
from sympy import symbols
x = symbols("x")
def alpha(x):
    return 2*x-5
def beta(x):
    return 1-x**2

print(alpha(beta(x)))
print(beta(alpha(x)))
```

## 4. Calculus

- Differentiate, or find $\frac{dy}{dx}$, the derivative of, $y=x^2+4x-3$
- Determine the gradient of the tangent of the curve, $y=4x^3-3x^4$, at the point $x=-\frac{1}{2}$. Plot the curve and its tangent.
- Integrate, $I=\int2x+5dx$.
- Determine the area under the curve, $y=4x^3-3x^4$, between $x=0.5$ and $x=1$ and above the $x$-axis. Plot the curve and shade the area. 

```python
# differentiation
from sympy import *
x = symbols("x")
diff(x**2+4*x-3, x)
```

```python
# find the gradient at x=-1/2
from sympy import *
x, y, c = symbols("x y c")
xval = -1/2
y = 4* x**3 - 3* x**4
dydx = diff(y,x)
grad = dydx.subs(x, xval)
yval = y.subs(x,xval)
int = (yval - grad*xval)
print("gradient_slope=", f"{grad:.4f}")
print("yval=", f"{yval:.4f}")
print("intercept=", f"{int:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols
x = np.linspace(-1, 1.5, 100)
y1 = 4* x**3 - 3* x**4
y2 = 12* x**2 - 12* x**3
y3 = 1.5625 + 4.5*x
plt.ylim(-7.5, 7.5)
plt.rcParams["font.size"] = 9
plt.plot(x, y1, label="$y=4x^3-3x^4$")
plt.plot(x, y2, label="$y{'}=12x^2-12x^3$")
plt.plot(x, y3,label="$y_{diff}=grad_{slope}*x+inter$")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```
